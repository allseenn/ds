```python
#       ()
fig = plt.figure()
ax = fig.add_subplot()
x = np.linspace(0, 30, 10)
y1 = x
y2 = x**2
ax.plot(x, y1, color='red', linestyle='--', label="legend_1")
ax.plot(x, y2, color='blue', linestyle='-', label="legend_2")

ax.legend()
ax.grid()
ax.set(title="Graph 1", xlabel="x", ylabel="y")
```

```python
#      
fig, axes = plt.subplots(2, 2)

x = np.linspace(0, 30, 10)
y1 = x
y2 = x**2
y3 = -x**2
y4 = np.sin(x)

for ax, y in zip(axes.flatten(), [y1, y2, y3, y4]):
    ax.plot(ax, y)
```

```python
##  
x = [1, 2, 3, 4, 5, 6, 7]
y1 = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
y2 = [7, 8, 8.5, 9, 9.5, 10, 10.5]
y3 = [10.5, 11, 11.5, 12, 12.5, 13, 13.5]
fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels=['y1', 'y2', 'y3'])
ax.legend(loc='upper left');
```

```python
#  
fig, ax = plt.subplots()
x = np.arange(0, 10.5, 0.5)
y = np.sin(x)
ax.scatter(x, y)
```

```python
#  
fig, ax = plt.subplots()
labels = [f'P{i}' for i in range(7)]
x = np.arange(len(labels))
counts = np.random.randint(3, 10, len(labels))
ax.bar(x, counts);
ax.set_xticks(x)
ax.set_xticklabels(labels);
```

```python
#   
labels = [f'P{i}' for i in range(5)]
g1 = [1, 2, 3, 4, 5]
g2 = [5, 4, 3, 2, 1]
width = 0.4
x = np.arange(len(labels))
fig, ax = plt.subplots()
ax.bar(x - width/2, g1, width, label='g1')
ax.bar(x + width/2, g2, width, label='g2')
ax.set_xticks(x)
ax.set_xticklabels(labels);
ax.grid()
ax.legend();
```

```python
#  
vals = [1, 2, 3, 4, 5]
labels = ["Ford", "Chevrolet", "BMW", "Toyota", "Nissan"]
fig, ax = plt.subplots()
ax.pie(vals, labels=labels);
```

```python
sns.relplot(x="horsepower", y="acceleration", size="cylinders", kind="scatter", data=mpg)
```

```python
sns.relplot(x="year", y="passengers", kind="line", data=flights)
```
