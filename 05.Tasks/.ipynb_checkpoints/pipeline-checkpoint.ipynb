Единственное с чем не согласен из задания - это с формулировкой четвёртой колонки Response, что в переводе означает "отклик", "ответ" или "ответная реакция", по контексту датасета можно перевести как "согласие", т.е. некая страховая фирма провела рекламную компанию, в результате которой клиент либо откликнулся либо нет, т.е. согласился купить страховку или отказался.
```python
def annot_plot(ax,w,h):                                    # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
         ax.annotate(f"{p.get_height() * 100 / data.shape[0]:.2f}%", (p.get_x() + p.get_width() / 2., p.get_height()),
         ha='center', va='center', fontsize=11, color='black', rotation=0, xytext=(0, 10),
         textcoords='offset points')             
def annot_plot_num(ax,w,h):                                    # function to add data to plot
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for p in ax.patches:
        ax.annotate('{0:.1f}'.format(p.get_height()), (p.get_x()+w, p.get_height()+h))
        
        
ax = sns.countplot(x='Response',data=data)
plt.ylabel('Total number of Response')
annot_plot(ax, 0.08,1)
plt.show()
```

### 4) обучить модели, и подобрать для каждой из них оптимальные гиперпараметры


```python
y = all_data_df['Response']
X = all_data_df.drop('Response', axis = 1)
```

```python
# Логистическая регрессия
from sklearn.linear_model import LogisticRegression

# Кросс-валидация
from sklearn.model_selection import cross_validate

log_reg = LogisticRegression(C = 0.8, random_state = 42, max_iter = 500)

cv_log_reg = cross_validate(log_reg, X, y, cv = 5, scoring = 'roc_auc')

print("Результаты Логистической регрессии:", sum(cv_log_reg['test_score']) / len(cv_log_reg['test_score']))
```

```python
# Решающее дерево
from sklearn.tree import DecisionTreeClassifier

dec_tree = DecisionTreeClassifier(random_state = 42) # Можно найти более оптимальные гиперпараметры

cv_dec_tree = cross_validate(dec_tree, X, y, cv = 5, scoring = 'roc_auc')

print("Результаты решающего дерева:", sum(cv_dec_tree['test_score']) / len(cv_dec_tree['test_score']))
```

```python
# Случайный лес
from sklearn.ensemble import RandomForestClassifier


ran_for = RandomForestClassifier(random_state = 42) # Можно найти более оптимальные гиперпараметры

cv_ran_for = cross_validate(ran_for, X, y, cv = 5, scoring = 'roc_auc')

print("Результаты решающего дерева:", sum(cv_ran_for['test_score']) / len(cv_ran_for['test_score']))
```

### 6) проинтерпретировать получившиеся результаты

```python
log_reg.fit(X, y)
print("Коэффициенты в линейной регрессии:", log_reg.coef_)
```

```python
dec_tree.fit(X, y)
print("Важность признаков решающего дерева", dec_tree.feature_importances_)
```

```python
import shap
# Загружаем JS шаблоны для визуализации
shap.initjs()

# Обязателен явный вызов метода fit
ran_for.fit(X, y)

print("Обучили модель, переходим к визуализации")
# Интерпретируем предсказания модели при помощи SHAP
explainer = shap.TreeExplainer(ran_for)
shap_values = explainer.shap_values(X)
```

```python
# Общий график
shap.summary_plot(shap_values, X, plot_type="bar")
```

На основе графика, видоно что двумя наиболее решающими фичами являются:
- Время, которое клиент сотрудничает со страховой компанией
- И его доход

Чтобы подтвердить данные выводы, построим график на основе релуьтатов обучения решающих деревьев:

```python
feature_imp = ran_for.feature_importances_.round(3)
ser_rank = pd.Series(feature_imp, index=X.columns).sort_values(ascending = False)

plt.figure(figsize=(12,7))
sns.barplot(x= ser_rank.values, y = ser_rank.index, palette='deep')
plt.xlabel('relative importance')
plt.show()
```

Выводы подтверждаются, страховку приобретают в основном люди, которые давно являются клиентами страховой компании и имеют достаточный для этого доход.
